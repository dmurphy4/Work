```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
```

```python
df = pd.read_csv('industry.csv',parse_dates = ['caldt'])
df.query ("caldt == '1963-07-01'")
```

```python
def ols_coef (x,formula) :
    reg = smf.ols(formula, data = x).fit()
    return reg.params

coefs = df.groupby('caldt').apply(ols_coef,'ret ~ 1 + beta + r12to2 + r36to13')
coefs
```

```python
def fm_summary(p):
    s = p.describe().T
    s['std_error'] = s['std']/np.sqrt(s['count'])
    s['tstat'] = s['mean']/s['std_error']
    return s[['mean', 'std_error', 'tstat']]

fm_summary(coefs)
```

```python
df = df.set_index(['caldt','industry'])
fm  = pd.fama_macbeth(y=df['ret'],x=df[['beta','r12to2','r36to13']])
fm
```
